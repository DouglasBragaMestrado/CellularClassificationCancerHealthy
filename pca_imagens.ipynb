```python
from math import exp
import numpy as np
from sklearn.decomposition import PCA
from pylab import imshow,gray,figure 
from PIL import Image, ImageOps 
import matplotlib.pyplot as plt 
import numpy as np
from pylab import imread,subplot,imshow,title,gray,figure,show,NullLocator 
%matplotlib inline
```

```python
#Abrindo a Imagem

A = Image.open("tmp/fei/img/1a.jpg") # open an image
#A = Image.open("tmp/fold_0/all/UID_11_10_1_all.bmp") # open an image
print(np.shape(A))
```

```python
A = np.mean(A,2)
full_pc = np.size(A,axis=1) # numbers of all the principal components r = len(A[:,0])/len(A[1])
i=1
b = len(A.T[0])
dist = []
full_pc =np.array([0, 0.33,0.66,])*b
full_pc = full_pc.astype(int)
i=1
perc = [0,33,66]
dist_list = []
plt.figure(figsize=(20,20))
```

```python
for numpc in full_pc:
    plt.subplot(2,2,i)
    pca = PCA(n_components=numpc)
    B=A.copy()
    X_pca = pca.fit_transform(B)
    reconstructed = pca.inverse_transform(X_pca)
    plt.title('PCA Components = %i (%i %%)'%(numpc,perc[i-1]),fontsize=20)
    dist_list.append(np.linalg.norm(A-reconstructed,'fro'))
    plt.imshow(reconstructed)
    plt.grid(False)
    i=i+1
plt.subplot(2,2,4)
imshow(A)
title('PCA FULL: '+str(b),fontsize=20) 
plt.grid(False)
gray()
show()
plt.tight_layout()
dist_list.append(0)
```
